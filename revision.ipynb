```python
a = [5,4,6,7,7,4,3,2,4]

smallest = a[0]
seconds = float('inf')

for i in range(1,len(a)):
    if a[i] < smallest:
        seconds = smallest
        smallest = a[i]
    
    elif a[i] > smallest and a[i] < seconds:
        seconds = a[i]

print(seconds)
```

```python
a = [1,1,1,4,3,4,5,6]

def isSorted(arr,n):
    for i in range(1,n):
        if arr[i-1] > arr[i]:
            return 'Not sorted'
    
    return 'Sorted'

isSorted(a,6)
       
```
