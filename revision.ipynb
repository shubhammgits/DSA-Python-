```python
def leftRotate(arr, n, k):
    k=k%n
    
    temp = []

    for i in range(k):
        temp.append(arr[i])

    for i in range(k,n):
        arr[i-k] = arr[i]

    for i in range(n-k, n):
        arr[i] = temp[i-(n-k)]

    return arr

arr = [1,2,3,4,5,6,7]
n = len(arr)
k = 3

leftRotate(arr,n,k)
```

```python
def reverse(arr,start,end):

    while start<end:
        arr[start], arr[end] = arr[end], arr[start]
        start+=1
        end-=1

    return arr

def leftRotate(arr, n, k):
    reverse(arr, 0, k-1)
    reverse(arr, k, n-1)
    reverse(arr, 0, n-1)

    return arr

arr = [1,2,3,4,5,6,7]
k=3
n=len(arr)

leftRotate(arr,n,k)
```

```python
#armstrong number : 

num = 153
n = len(str(num))
sum=0
digits = num
while digits>0:
    lastDigit = digits%10
    sum = sum + (lastDigit)**n
    digits = int(digits//10)


if sum == num:
    print(f'{num} is an Armstrong number!')
```

```python
# check prime number 

def isPrime(num):
    if num<2:
        return 'Non Prime'

    for i in range(2, int(num**0.5)+1):
        if num%i == 0:
            return 'Non Prime'
    
    return 'Prime'

print(isPrime(1))
print(isPrime(5))
print(isPrime(2))
print(isPrime(12))
print(isPrime(122))
print(isPrime(19))
```

```python
# reverse

digit = 123
copy = digit
result = 0

while copy>0:
    lastDigit = copy%10
    result = result*10 + lastDigit
    copy = copy//10

print(result)
```

```python
# palindrome

def isPalindrome(num):
    copy = num
    result = 0

    while copy>0:
        lastDigit = copy%10
        result = result*10 + lastDigit
        copy = copy//10

    if result == num:
        print(f'{num} is a Palindrome number!')

    else:
        print(f'{num} is not a Palindrome number!')

isPalindrome(121)
isPalindrome(123)
```
