```python
def leftRotate(arr, n, k):
    k=k%n
    
    temp = []

    for i in range(k):
        temp.append(arr[i])

    for i in range(k,n):
        arr[i-k] = arr[i]

    for i in range(n-k, n):
        arr[i] = temp[i-(n-k)]

    return arr

arr = [1,2,3,4,5,6,7]
n = len(arr)
k = 3

leftRotate(arr,n,k)
```

```python
def reverse(arr,start,end):

    while start<end:
        arr[start], arr[end] = arr[end], arr[start]
        start+=1
        end-=1

    return arr

def leftRotate(arr, n, k):
    reverse(arr, 0, k-1)
    reverse(arr, k, n-1)
    reverse(arr, 0, n-1)

    return arr

arr = [1,2,3,4,5,6,7]
k=3
n=len(arr)

leftRotate(arr,n,k)
```
