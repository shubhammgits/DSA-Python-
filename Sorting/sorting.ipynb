## Selection Sort

```python
def selectionSort(nums):
    n=len(nums)

    for i in range(n-1):
        mini=i
        for j in range(i+1,n):
            if nums[j]<nums[mini]:
                mini=j

        nums[i],nums[mini] = nums[mini],nums[i]



nums=[13,46,24,52,30,9]
selectionSort(nums)
print(nums)
```

## Bubble Sort

```python
def bubbleSort(arr):
    n=len(arr)

    for i in range(n-1,0,-1):
        didSwap=0
        for j in range(0,i):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                didSwap=1

        if didSwap==0:
            break

arr=[13,46,24,52,20,9]
bubbleSort(arr)
print(arr)
```
