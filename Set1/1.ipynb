```python
# prime numbers between a range:

a = int(input('Enter the starting number : '))
b = int(input('Enter the last number : '))

for num in range(a,b+1):
    if num<2:
        continue
    isPrime = True
    for i in range(2,int(num**0.5)+1):
        if num%i == 0:
            isPrime = False
            break

    if isPrime:
        print(num)
```

```python
#armstrong number : 

num = input('Enter the number : ')
sum = 0
l = len(num)

intnum = int(num)

while intnum>0:
    digit = intnum%10
    sum = sum + digit**l
    intnum = intnum//10

if sum == int(num) :
    print('Armstrong')
else:
    print('Not armstrong')
```

```python
num = 10
L=[]
for i in range(num+1):
    isPrime = True
    if i<2:
        continue
    for j in range(2,int(i**0.5)+1):
        if i%j == 0:
            isPrime = False
            break
    if isPrime:
        L.append(i)

for nums in L:
    other = num-nums

    if other in L and nums<=other:
        print(num, '=', nums, '+', other)
        

    
```

```python
# program to replace all 0's with 1 in an integer

digit = 1010
L = []

while digit>0:
    n = digit%10
    L.append(n)
    digit = digit//10

print(L)

for i in range(len(L)):

    if L[i] == 0:
        L[i] == 1

print(L)
```
