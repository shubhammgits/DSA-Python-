```python
# prime numbers between a range:

a = int(input('Enter the starting number : '))
b = int(input('Enter the last number : '))

for num in range(a,b+1):
    if num<2:
        continue
    isPrime = True
    for i in range(2,int(num**0.5)):
        if num%i == 0:
            isPrime = False
            break

    if isPrime:
        print(num)
```

```python
#armstrong number : 

num = input('Enter the number : ')
sum = 0
l = len(num)

intnum = int(num)

while intnum>0:
    digit = intnum%10
    sum = sum + digit**l
    intnum = intnum//10

if sum == int(num) :
    print('Armstrong')
else:
    print('Not armstrong')
```
