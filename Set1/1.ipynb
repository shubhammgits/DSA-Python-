```python
num = 10
L=[]
for i in range(num+1):
    isPrime = True
    if i<2:
        continue
    for j in range(2,int(i**0.5)+1):
        if i%j == 0:
            isPrime = False
            break
    if isPrime:
        L.append(i)

for nums in L:
    other = num-nums

    if other in L and nums<=other:
        print(num, '=', nums, '+', other)
        

    
```

```python
# program to replace all 0's with 1 in an integer

digit = 1010

place=1
result=0

while digit>0:
    n = digit%10

    if n==0:
        n = 1
    
    result = result + n*place

    place = place*10
    digit = digit//10
print(result)
```

```python
#pyramid pattern using stars:
n=4
for i in range(0,n):

    #space
    for j in range(0,n-i-1):
        print(' ',end='')
    
    #stars
    for j in range(0,1+2*i):
        print('*', end='')
    
    #space
    for j in range(0,n-i-1):
        print(' ', end='')
    
    print()
```

```python
#pyramid pattern using numbers

n=4
for i in range(1,n+1):

    #space
    for j in range(0,n-i):
        print(' ', end='')

    #number
    for j in range(0,2*i-1):
        print(i, end='')

    #space
    for j in range(0,n-1):
        print(' ', end='')

    print()
```

```python
L = [1,2,3,4,5]
n = len(L)
for i in range(0,n//2):
    last = n-i-1
    temp = L[i]

    L[i] = L[last]
    L[last] = temp

print(L)



```
