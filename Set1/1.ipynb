```python
# prime numbers between a range:

def printPrimeNum(a,b):
    L=[]
    for i in range(a,b+1):
        if i<=2:
            L.append(i)
        for j in range(2,int(i**0.5)):
            if i%j == 0:
                break
            else:
                L.append(i)
    return L

printPrimeNum(1,10)
```
