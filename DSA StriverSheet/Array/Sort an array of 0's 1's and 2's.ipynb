## Sort an array of 0 1 and 2

```python
#better:
arr = [0,1,2,0,1,2,1,2,0,0,0,1]
n=len(arr)

count0,count1,count2=0,0,0

for i in range(n):
    if arr[i] == 0:
        count0+=1
    
    elif arr[i] == 1:
        count1+=1
    
    else:
        count2+=1

print(count0,count1,count2)

for i in range(count0):
    arr[i]=0

for i in range(count0, count0+count1):
    arr[i]=1

for i in range(count0+count1, n):
    arr[i]=2

print(arr)
```

```python
#dutch national flag algo (optimal)

def func(arr):
    n=len(arr)
    low,mid,high=0,0,n-1

    while (mid<=high):

        if arr[mid] == 0:
            arr[mid],arr[low] = arr[low],arr[mid]
            mid+=1
            low+=1

        elif arr[mid] == 1:
            mid+=1

        else:
            arr[mid],arr[high] = arr[high],arr[mid]
            high-=1

arr = [0,1,2,0,1,2,1,2,0,0,0,1]
func(arr)
print(arr)
```
