### Rotate an array by K places

```python
#left rotate an array by one place

a = [1,2,3,4,5]
temp = a[0]
n = len(a)
for i in range(1,n):
    a[i-1] = a[i]
    
a[n-1]=temp
print(a)
```

```python
#left rotate array by k places

#brute

a = [1,2,3,4,5,6,7]
n = len(a)

k = 3
tem=[]

for i in range(k):
    tem.append(a[i])

# print(tem)

for i in range(k,n):
    a[i-k] = a[i]


for i in range(n-k,n):
    a[i] = temp[i-(n-k)]

print(a)
```

```python
#better:

def leftRotate(arr,n,k):

    k = k%n

    temp = []
    for i in range(k):
        temp.append(arr[i])

    for i in range(k,n):
        arr[i-k] = arr[i]

    for i in range(n-k,n):
        arr[i] = temp[i-(n-k)]

    return arr


arr = [1,2,3,4,5,6,7]
n = len(arr)
k = 3
leftRotate(arr,n,k)
```

```python
#optimal:

def reverse(arr,start,end):
    while start<end:
        arr[start], arr[end] = arr[end], arr[start]
        start+=1
        end-=1

def leftRotate(arr,n,k):
    reverse(arr, 0, k-1)
    reverse(arr, k, n-1)
    reverse(arr, 0, n-1)

    return arr

arr=[1,2,3,4,5,6,7]
n=len(arr)
k=3
leftRotate(arr,n,k)
```

```python
#reverse an array:

def reverse(arr, start, end):
    while start<end:
        arr[start],arr[end] = arr[end], arr[start]
        start+=1
        end-=1

    return arr

arr = [1,2,3,4,5]
reverse(arr,0,len(arr)-1)
```

```python
#right rotation

def reverse(arr,start,end):
    while start<end:
        arr[start], arr[end] = arr[end], arr[start]
        start+=1
        end-=1

def rightRotate(arr,n,k):
    reverse(arr,0,n-1)
    reverse(arr, 0, k-1)
    reverse(arr, k, n-1)

    return arr

arr= [1,2,3,4,5,6,7]
n=len(arr)
k=3

rightRotate(arr,n,k)
```

### Move all zeroes to the end of an array

```python
a = [1,0,2,3,2,0,0,4,5,1]
res=[]
zeroes=[]

for i in a:
    if i!=0:
        res.append(i)
    if i==0:
        zeroes.append(i)

for i in range(len(zeroes)):
    res.append(zeroes[i])

print(res)
```
