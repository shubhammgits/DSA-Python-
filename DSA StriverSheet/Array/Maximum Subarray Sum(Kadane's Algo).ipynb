## Maaximum Subarray Sum

```python
#brute:

arr = [-2,-3,4,-1,-2,1,5,-3]
n=len(arr)
maxi=float('-inf')

for i in range(n):
    for j in range(i,n):
        sum=0
        for k in range(i,j+1):
            sum+=arr[k]

        maxi = max(sum,maxi)

print(maxi)
```

```python
#better:


arr = [-2,-3,4,-1,-2,1,5,-3]
maxi=float('-inf')
n = len(arr)

for i in range(n):
    currsum=0
    for j in range(i,n):
        currsum+=arr[j]

        maxi = max(currsum,maxi)

print(maxi)
```

```python
#optimal: kadane's

sum=0
n=len(arr)
maxi = float('-inf')

for i in range(n):
    sum+=arr[i]

    if sum>maxi:
        maxi=sum

    if sum<0:
        sum=0


print(maxi)
```

```python
#to print the subarray too:

n=len(arr)
maxi =float('-inf')
currSum=0
start=0
ansStart=-1
ansEnd=-1

for i in range(n):

    if currSum == 0:
        start=i

    currSum += arr[i]

    if currSum>maxi:
        maxi=currSum
        ansStart = start
        ansEnd = i

    if currSum<0:
        currSum=0

print('Maximum Sum : ',maxi)
print('Subarray : ',end=' ')
l=[]
for i in range(ansStart,ansEnd+1):
    l.append(arr[i])

print(l)
```
