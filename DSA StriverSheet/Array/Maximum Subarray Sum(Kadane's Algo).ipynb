## Maaximum Subarray Sum

```python
#brute:

arr = [-2,-3,4,-1,-2,1,5,-3]
n=len(arr)
maxi=float('-inf')

for i in range(n):
    for j in range(i,n):
        sum=0
        for k in range(i,j+1):
            sum+=arr[k]

        maxi = max(sum,maxi)

print(maxi)
```

```python
#better:


arr = [-2,-3,4,-1,-2,1,5,-3]
maxi=float('-inf')
n = len(arr)

for i in range(n):
    currsum=0
    for j in range(i,n):
        currsum+=arr[j]

        maxi = max(currsum,maxi)

print(maxi)
```

```python
#optimal: kadane's

sum=0
n=len(arr)
maxi = float('-inf')

for i in range(n):
    sum+=arr[i]

    if sum>maxi:
        maxi=sum

    if sum<0:
        sum=0


print(maxi)
```
