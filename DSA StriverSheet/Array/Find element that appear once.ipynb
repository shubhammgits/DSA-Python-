### FInd the missing number in a array

```python
#optimal

a = [1,2,4,5]
n=5

sumN = 0
sumArray = 0

for i in range(1,n+1):
    sumN+=i

print(sumN)

for nums in a:
    sumArray += nums

missingNumber = sumN-sumArray
print(missingNumber)

```

```python
#more optimal better for larger n values like 10^5

a = [1,2,4,5]
N=5

xor1 = 0
xor2 = 0

for i in range(1,N+1):
    xor1 = xor1^i

for i in range(N-1):
    xor2 = xor2^a[i]

print(xor1^xor2)

```

### Maximum consecutive ones

```python
#optimal sol is staright forward so no brute better

a = [1,1,0,1,1,1,0,1,1]

count,maxi=0,0

for i in range(len(a)):
    if a[i] == 1:
        count+=1
        maxi = max(maxi,count)

    else:
        count=0

print(maxi)

```

### Find the number that appears once, and others twice

```python
#brute

a = [1,1,2,3,3,4,4]

for nums in a:
    count=0
    for j in range(len(a)):
        if a[j] == nums:
            count+=1

    if count == 1:
        print(nums)
        
```
