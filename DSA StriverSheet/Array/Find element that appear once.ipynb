### FInd the missing number in a array

```python
#optimal

a = [1,2,4,5]
n=5

sumN = 0
sumArray = 0

for i in range(1,n+1):
    sumN+=i

print(sumN)

for nums in a:
    sumArray += nums

missingNumber = sumN-sumArray
print(missingNumber)

```
