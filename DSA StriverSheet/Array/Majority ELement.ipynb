## Majority ELement

```python
#brute:

arr = [2,2,3,3,1,2,2]
n=len(arr)

def brute(arr,n):
    for i in range(n):
        count=0
        for j in range(n):
            if arr[i]==arr[j]:
                count+=1

        if count>n/2:
            return arr[i]
        
        else:
            return -1

brute(arr,n)
```
