## Majority ELement

```python
#brute:

arr = [2,2,3,3,1,2,2]
n=len(arr)

def brute(arr,n):
    for i in range(n):
        count=0
        for j in range(n):
            if arr[i]==arr[j]:
                count+=1

        if count>n/2:
            return arr[i]
        
        else:
            return -1

brute(arr,n)
```

```python
#better: hashing

arr = [2,2,3,3,1,2,2,]
n=len(arr)
freq = {}

for nums in arr:
    if nums in freq:
        freq[nums]+=1
    else:
        freq[nums]=1

for key in freq:
    if freq[key]>n/2:
        print(key)
```

```python
#moor's voting algo(optimal solution):

arr = [7,7,5,7,5,1,5,7,5,5,7,7,5,5,5,5]
count=0
elem=0
n=len(arr)

for i in range(n):
    if count==0:
        count=1
        elem = arr[i]

    elif arr[i] == elem:
        count+=1

    else:
        count-=1

verifyCount=0
for i in range(n):
    if arr[i] == elem:
        verifyCount+=1

if verifyCount>n//2:
    print(elem)
else:
    print(-1)
```
