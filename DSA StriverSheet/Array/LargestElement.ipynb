### Largest Element in an Array

### Brute force : sorting then printing the last element (O(nlogn))

## Optimal : O(n)

```python
l = [7,3,2,1,5,8,2]
```

```python
largest = l[0]
for i in range(len(l)):
    if l[i] > largest:
        largest=l[i]
print(f'Largest element is {largest}')
```

### Second Largest Element in an Array

```python
l = [1,2,4,7,7,5]
```

```python
#brute force : firt sort the array and largest = l[n-1] and then :

# for i in range(len(l)-2, 0, -1):
#    if l[i]!=largest:
#          secondL = l[i]
#          break
```

```python
# better solution:

#first pass: find the largest:

arr = [1,2,4,7,7,5]

largest = arr[0]
for i in range(len(arr)):
    if arr[i] > largest:
        largest = arr[i]


#second pass:

secondL = -1
for i in range(len(arr)):
    if arr[i] > secondL and arr[i] != largest:
        secondL = arr[i]

print(f'Second Largest element is {secondL}')

```

```python
#optimal solution: (assuming array doesnt contain any negative number and if contains then take secondL = INT_MIN (in above approaches too))


#find the second largest and second smallest element in an array


def secondLargest(arr,n):
    largest = arr[0]
    slargest = float('inf')

    for i in range(1,n):
        if arr[i]>largest:
            slargest = largest
            largest = arr[i]

        elif arr[i] < largest and arr[i] > slargest:
            slargest = arr[i]

    return slargest

def secondSmallest(arr,n):
    smallest = arr[0]
    secondsm = float('inf')

    for i in range(1, n):

        if arr[i] < smallest:
            secondsm = smallest
            smallest = arr[i]

        elif arr[i] > smallest and arr[i] < secondsm:
            secondsm = arr[i]
    
    return secondsm

arr = [1,2,4,7,7,5]
n = len(arr)

slargest = secondLargest(arr,n)
ssmallest = secondSmallest(arr,n)

print(slargest, ssmallest)
```

### Check if the array is sorted

```python
arr = [0,0,1,-1,3,4]

def isSorted(arr,n):
    for i in range(1,n):

        if arr[i] < arr[i-1]:
            return 'Not Sorted'
    
    return 'Sorted Array'

isSorted(arr, len(arr))
```
