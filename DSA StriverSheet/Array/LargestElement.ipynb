### Largest Element in an Array

### Brute force : sorting then printing the last element (O(nlogn))

## Optimal : O(n)

```python
l = [7,3,2,1,5,8,2]
```

```python
largest = l[0]
for i in range(len(l)):
    if l[i] > largest:
        largest=l[i]
print(f'Largest element is {largest}')
```
