## Two Sum Problem

```python
#brute

arr=[2,6,5,8,11]
target=14
n=len(arr)

for i in range(0,n):
    for j in range(i+1,n):
        if i==j:
            continue

        if arr[i]+arr[j]==target:
            print(i,j)
            
```

```python
#better : hashing

arr = [2,6,5,8,11]
n=len(arr)
target=14

def hash(arr,n,target):
    freq={}
    for i in range(len(arr)):
        first = arr[i]
        more = target - first

        if more in freq:
            return 'YES'
        freq[first]=i

    return 'NO' 

print(hash(arr,n,target))
```

```python
#better : hashing to print the indices 

def twoSum(arr,n,target):
    freq={}
    for i in range(n):
        more=target-arr[i]

        if more in freq:
            return [freq[more],i]

        freq[arr[i]]=i
    
    return []

twoSum(arr,n,target)
```

```python
#optimal solution for variety 1 that is only return yes or no

def twoSum(arr,n,target):
    left,right=0,n-1

    arr.sort()
    while left<right:
        currsum=arr[left]+arr[right]
        if currsum == target:
            return 'YES'
        
        elif currsum>target:
            right-=1
        
        else:
            left+=1

    return 'NO'


arr = [2,6,5,8,11]
target=14
n=len(arr)
twoSum(arr,n,target)
```
