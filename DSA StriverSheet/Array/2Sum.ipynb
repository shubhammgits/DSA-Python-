## Two Sum Problem

```python
#brute

arr=[2,6,5,8,11]
target=14
n=len(arr)

for i in range(0,n):
    for j in range(i+1,n):
        if i==j:
            continue

        if arr[i]+arr[j]==target:
            print(i,j)
            
```

```python
#better : hashing

arr = [2,6,5,8,11]
n=len(arr)
target=14

def hash(arr,n,target):
    freq={}
    for i in range(len(arr)):
        first = arr[i]
        more = target - first

        if more in freq:
            return 'YES'
        freq[first]=i

    return 'NO' 

print(hash(arr,n,target))
```
