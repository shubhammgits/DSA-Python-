### Longest subarray with sum k

```python
#brute: generate all the subarrays and caluclate the sum

a = [1,2,3,1,1,1,1,4,2,3]
target = 3
longestLength=0

for i in range(len(a)):
    for j in range(i,len(a)):
        currsum=0
        for k in range(i,j+1):
            currsum+=a[k]
        
        if currsum==target:
            longestLength = max(longestLength, j-i+1)

print(longestLength)
```

```python
#improved brute
a = [1,2,3,1,1,1,1,4,2,3]
target=3
longestLength=0

for i in range(len(a)):
    currSum=0
    for j in range(i,len(a)):
        currSum += a[j]

        if currSum == target:
            longestLength = max(longestLength,j-i+1)

print(longestLength)
```

```python
a = [1,2,3,1,1,1,1,4,2,3]
k=3

prefixSum=0
longestLength=0
prefixMap={}

for i in range(len(a)):
    prefixSum += a[i]

    #if the subarray starts from index 0
    if prefixSum == k:
        longestLength = i+1

    remaining = prefixSum-k

    if remaining in prefixMap:
        length = i-prefixMap[remaining]
        longestLength = max(longestLength,length)

    if prefixSum not in prefixMap:
        prefixMap[prefixSum] = i

print(longestLength)
```
