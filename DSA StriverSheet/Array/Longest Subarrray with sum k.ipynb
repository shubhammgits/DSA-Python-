### Longest subarray with sum k

```python
#brute: generate all the subarrays and caluclate the sum

a = [1,2,3,1,1,1,1,4,2,3]
target = 3

max_len = 0

for i in range(len(a)):
    for j in range(i, len(a)):
        curr_sum = 0

        for x in range(i, j + 1):
            curr_sum += a[x]

        if curr_sum == target:
            max_len = max(max_len, j - i + 1)

print(max_len)
```
