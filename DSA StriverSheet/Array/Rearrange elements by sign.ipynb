## Rearrange elements by sign (same order, alternate position)

```python
#brute:

arr = [3,1,-2,-5,2,-4]

n = len(arr)
pos =[]
neg = []

for i in range(n):
    if arr[i]>0:
        pos.append(arr[i])
    else:
        neg.append(arr[i])

print(pos)
print(neg)

for i in range(n//2):
    arr[2*i] = pos[i]
    arr[2*i+1] = neg[i]

print(arr)
```

```python
arr = [3,1,-2,-5,2,-4]
n = len(arr)
ans = [0]*n

posInd = 0
negInd = 1

for i in range(n):
    if arr[i]<0:
        ans[negInd] = arr[i]
        negInd += 2

    else:
        ans[posInd] = arr[i]
        posInd += 2

print(ans)
```
