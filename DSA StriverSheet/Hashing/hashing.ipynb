```python
#brute force to fine number of times an element exist in an array:

a = [1,2,1,3,1,4,2,3]

def func(arr,num):
    count=0

    for elem in arr:
        if elem == num:
            count+=1

    return count
```

```python
print(func(a,1))
print(func(a,2))
```

```python
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

for num in m:
    count=0
    for x in n:
        if num==x:
            count+=1

    print(num ,'=', count)
```

```python
#using hashlist: (check constraint in notebook first)

n = [5,3,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

hashlist = [0]*11

for num in n:
    hashlist[num]+=1

for num in m:
    if num<1 or num>10:
        print(num,'=',0)
    
    else:
        print(num,'=',hashlist[num])
```

```python
#using dictionary

n = [5,3,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

freq = {}

for num in n:
    if num in freq:
        freq[num]+=1

    else:
        freq[num]=1

print(freq)

for num in m:
    if num in freq:
        print(num, '=', freq[num])
    else:
        print(num,'=',0)

```

```python
#character hashing

s = 'azyxyyzaaaa'
q = ['d','a','y','x']
hashlist = [0]*26

for ch in s:
    asciival = ord(ch)
    index = asciival - 97
    hashlist[index]+=1

for ch in q:
    asciival = ord(ch)
    index = asciival - 97
    print(ch,'=',hashlist[index])
```

```python
#using dictionary:

s = 'AkbNh8NdgfLkI0@9%'
q = 'AB@bNDgL0'

freq = {}

for elem in s:
    if elem in freq:
        freq[elem]+=1

    else:
        freq[elem]=1

for elem in q:
    if elem in freq:
        print(elem,'=',freq[elem])
    else:
        print(elem,'=',0)
```

```python
#Print only the number of distinct elements.

arr = [1,2,1,3,4,3,5]
freq={}
count=0
for nums in arr:
    if nums in freq:
        freq[nums]+=1
    else:
        freq[nums]=1
        count+=1

print(count)
```

```python
#print only the duplicate items

arr = [1,2,3,2,1,5,6,6]

freq={}
for nums in arr:
    if nums in freq:
        freq[nums]+=1
    else:
        freq[nums]=1

for key in freq:
    if freq[key] > 1:
        print(key)
```

```python
#Print the first repeating element.

arr = [10,5,3,4,3,5,6]

freq = {}
for nums in arr:
    if nums in freq:
        freq[nums]+=1
    else:
        freq[nums]=1


for nums in arr:
    if freq[nums]>1:
        print(nums)
        break
```

```python
#Print the first non-repeating element.

arr = [4,5,1,2,0,4]

freq = {}

for nums in arr:
    if nums in freq:
        freq[nums]+=1
    else:
        freq[nums]=1

for nums in arr:
    if freq[nums]==1:
        print(nums)
        break
```

```python
#Print the array after removing duplicates while preserving the original order

arr = [1,2,2,3,1,4,5,4]
freq = {}

for nums in arr:
    if nums not in freq:
        print(nums,end=' ')
        freq[nums]=1
```

```python
def isDuplicates(arr):
    freq={}
    for nums in arr:
        if nums in freq:
            freq[nums]+=1
        else:
            freq[nums]=1

    for nums in arr:
        if freq[nums]>1:
            return True
        else:
            return False

print(isDuplicates([1,2,3,4,5]))
print(isDuplicates([1,2,3,2]))
```
