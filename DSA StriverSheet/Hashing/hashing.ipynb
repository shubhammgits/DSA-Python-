## HASHING

```python
#brute force to fine number of times an element exist in an array:

a = [1,2,1,3,1,4,2,3]

def func(arr,num):
    count=0

    for elem in arr:
        if elem == num:
            count+=1

    return count
```

```python
print(func(a,1))
print(func(a,2))
```

```python
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

for num in m:
    count=0
    for x in n:
        if num==x:
            count+=1

    print(num ,'=', count)
```

```python
#using hashlist: (check constraint in notebook first)

n = [5,3,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

hashlist = [0]*11

for num in n:
    hashlist[num]+=1

for num in m:
    if num<1 or num>10:
        print(num,'=',0)
    
    else:
        print(num,'=',hashlist[num])
```
