## Recursion

```python
#print 'hello' 4 times using recusrion
#head recusrion

def func(count):
    if count==4:
        return
    print('hello')
    func(count+1)

func(0)
```

```python
#tail recusrion

def func(count):
    if count==4:
        return
    func(count+1)
    print('hello')

func(0)
```

```python
#recusrion using parameters

def func(x,n):

    #base condition
    if n==0:
        return

    print(x)
    func(x,n-1)

func(15,4)
```

```python
#print 1 to n using recursion

def func(i,n):

    if i>n:
        return

    print(i)
    func(i+1,n)

func(1,5)
```

```python
#print 1 to n using tail recursion:

def func(i,n):

    if i>n:
        return

    func(i+1,n)
    print(i)

func(1,5)  #n to 1 will be printed!!
```

```python
# again try printing 1 to n using backtrqaking that is tail recursion

def func(n):
    
    if n==0:
        return
    func(n-1)
    print(n)

func(5)
```

```python
#sum 1 to n:

def func(sum,i,n):
    if i>n:
        print(sum)
        return

    func(sum+i,i+1,n)

func(0,1,4)
```

```python
#functional recusrion

def func(n):

    if n==1:
        return 1
    return n+func(n-1)

func(10)
```

```python
#factorial

def func(n):

    if n==0 or n==1:
        return 1

    return n*func(n-1)

func(4)
```

```python
#reverse an array:

arr = [5,7,3,2]
n=len(arr)
i,j=0,n-1

while i<j:
    arr[i],arr[j]=arr[j],arr[i]
    i+=1
    j-=1

print(arr)

```

```python
#using recursion:

arr = [5,9,8,3,6,7,1,4,2]
left=1
right=5

def func(arr,left,right):
    if left>right:
        return 

    arr[left],arr[right] = arr[right],arr[left]
    func(arr,left+1,right-1)
    return arr

func(arr,left,right)
```

```python
#check if a string is palindrome or not
s = 'ANBCDDCBNA'


def func(s):
    n=len(s)
    i,j=0,n-1
    while i<j:
        if s[i]!=s[j]:
            return False
    
        i+=1
        j-=1
    
    return True


func(s)
```
