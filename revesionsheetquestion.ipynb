#### 1. check if a number is prime or not

```python
def isPrime(num):
    if num<2:
        return 'Non Prime'

    for i in range(2, int(num**0.5)+1):
        if num%i == 0:
            return 'Non Prime'
    
    return 'Prime'
```

#### 2. reverse a number

```python
digit = 123
copy = digit
result = 0

while copy>0:
    lastDigit = copy%10
    result = result*10 + lastDigit
    copy = copy//10

print(result)
```

#### 3. check if a number is palindrome or not

```python
def isPalindrome(num):
    copy = num
    result = 0

    while copy>0:
        lastDigit = copy%10
        result = result*10 + lastDigit
        copy = copy//10

    if result == num:
        print(f'{num} is a Palindrome number!')

    else:
        print(f'{num} is not a Palindrome number!')

isPalindrome(121)
isPalindrome(123)
```

#### 4. Armstrong number

```python
digit = 153
n = len(str(digit))
result = 0
copy = digit

while copy>0:
    lastDigit = copy%10
    result = result + lastDigit**n
    copy = copy//10

if result == digit:
    print(f'{digit} is an armstrong number!')

else:
    print(f'{digit} is not an armstrong number!')
```
