```python
import  ctypes
```

```python
class MeraList:

    def __init__(self):
        self.size = 1        #size will tell max kitne items store kr skte hai
        self.n = 0           #n tell right now how much items are stored


         #create a ctype array when size = self.size

        self.A = self.__make_array(self.size)
    
    def __len__(self):
        return self.n

    def __str__(self):
        result = ''
        for i in range(self.n):
            result = result + str(self.A[i]) + ','
        
        return result


    def append(self,item):
        if self.n == self.size:
            #resize
            self.__resize(self.size*2)
        
        #append
        self.A[self.n] = item
        self.n += 1

    def __resize(self,new_capacity):
    #create a new array with new capacity
        B = self.__make_array(new_capacity)
        self.size = new_capacity

        #copy the content of A to B:
        for i in range(self.n):
            B[i] = self.A[i]
        
        #reassign A:
        self.A = B

    def __make_array(self,capacity):
        return (capacity*ctypes.py_object)()  #this code creates a ctype static,referential array with size capacity

```

```python
L = MeraList()
```

```python
L.append('Hello')
L.append(3.4)
L.append(True)
L.append(100)
```
