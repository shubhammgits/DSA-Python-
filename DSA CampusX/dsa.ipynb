### Python lists are dynamic arrays

```python
import sys
```

```python
L = []
sys.getsizeof(L)
```

```python
items = ['Hello', 1, 3.5, True, 'World', 78]
for item in items:
    print(L, sys.getsizeof(L))
    L.append(item)
```
